# 1. Creation and Simulation of a Synthetic Karst System with MODFLOW-CFP

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os

import CFPy as cfpy
import flopy as flopy
import pykasso as pk
```

## 1.1 Create MODFLOW Model (FloPy)

We use pyKasso here to generate a node / conduit network for MODFLOW-CFP. \
CFPy is used to generate the CFP-related input files and FloPy is used to generate the remaining input files for MODFLOW.

```python
""" create model object """

# change active directory to model directory
path = os.getcwd() + "\\modflow\\"
os.chdir(path)

modelname = "CFPy_example_01"
# using the CFPv2 here, which does support CADS
# to change, type in the correct CFP version for exe_name
mf = flopy.modflow.Modflow(modelname, exe_name="cfpv2")
```

```python
""" define general domain characteristics """

# NOTE: n_rows, n_cols, n_lays needs to be the same as specified in the .nbr-file!

# matrix dimensions
# number of rows in the domain, int
n_rows = 20

# number of columns in the domain, int
n_cols = 50

# number of layers in the domain, int
n_lays = 2

#width of each cell along rows
delr = 500.0

#width of each cell along columns
delc = 500.0

""" hydraulic characteristics """

# units: [L: m], [T: d] --> specify time_unit = 4 below!!
# units: [L: m], [T: h] --> specify time_unit = 3 below!!

#horizontal hydraulic conductivity of the porous matrix [L/T]
hk = 0.1

#vertical hydraulic conductivity [L/T]
vka = hk / 2
#specific storage [1/L]
ss = 1e-4
#specific yield [-]
sy = 0.001

""" time discretization """

# using one stress period for every day of the precipitation time series

# time units: 3=hours, 4=days
time_unit = 4

# define number of stress periods to be equal to the number of days in the recharge time series
n_pers = 500

# length (number of steps) of stress periods
# this variable assumes a uniform time discretization, i.e. all stress periods
# have the same number of time steps except the first stationary time step
perlen = 1

# number of time steps in each stress period (integer array)
# e.g., [1, 50, 50, 70, 33, ...]
n_stps = np.ones(n_pers) * perlen
n_stps = n_stps.tolist()
# ensure that fist stress period (stationary) has length 1
n_stps[0] = 1.0

# list with "False" everywhere except 0th element (which is "True")
#     to specify steady state ("True") or transient ("False")
#     stress periods
steady = np.zeros(n_pers, dtype=np.bool)
# ensure stationary first stress period
steady[0] = 1

#array of stress period numbers, 1-based indexing
spers = np.arange(1,n_pers + 1)

""" boundary conditions """

# head at outflow / spring node (last column and center row of domain)
node_head_bc = 200.0
# matrix head on the left side (first column)
H_left = 230.0

# matrix starting head on the right side (last column)
H_right = 200.0

# solver tolerances
# modflow general tolerance
mftol = 1e-3
# modflow relaxation parameter
mfrelax = 0.99
# CFP general tolerance
cfptol = 1e-5
# CFP relaxation parameter
cfprelax = 0.99
```

```python
""" generate some layer elevation data """

# set seed for reproducibility
np.random.seed(11)

# create random elevations
mtop = np.random.normal(300, 20, (20, 50))
l1bot = np.random.normal(150, 20, (20, 50))
l2bot = np.zeros((20, 50))

lay_elevs = [mtop, l1bot, l2bot]

# plot
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 4), dpi=200)
for i in range(3):
    im = ax[i].imshow(lay_elevs[i])
    plt.colorbar(im, ax=ax[i], shrink=0.2)
```

## 1.2 Generate pyKasso network

```python
catchment = pk.SKS(yaml_settings_file=path + "settings.yaml")

# get data for geology, faluts, and fractures as arrays
geology = catchment.get_geology()
faults = catchment.get_faults()
fractures = catchment.get_fractures()

# show maps of geology, faults, and fractures
catchment.show_catchment(data='geology')
catchment.show_catchment(data='faults')
catchment.show_catchment(data='fractures')
```

```python
# compute karst networks from the given information
catchment.compute_karst_network()
```

```python
# generate elevation data
# NOTE: elevation data has to have the same shape as the node network array!
shp = np.array(catchment.karst_simulations[-1].maps['karst'][-1]).shape
np.random.seed(10)
elevs = np.random.normal(30, 5, shp)

# validate the network from pyKasso
validator = cfpy.preprocessing.pyKassoValidator(network=catchment, elevations=elevs)
valid_network = validator.validate_network()

# plot the network
im = plt.imshow(valid_network)
plt.colorbar(im, shrink=0.3)
```

```python
# export the network
# the exported information can directly be included in the .nbr-file as input for CFPy
# notes on how to use the generated data with CFPy is given at the end of the notebook
# per default, the network is exported to "CFPy_exported_network_for_NBR.txt" in the active directory
validator.export_network()
```

## 1.3 Create CFP Input Files
- .nam-file of MODFLOW model gets updated at the end
- relevant CFP-files are created on the way

```python
validator.generate_nbr(
    path = path,
    nrows = n_rows,
    ncols = n_cols,
    nlays = n_lays,
    nplanes = 1,
    layer_elevations = lay_elevs
)
```

### CFPy - NBR

```python
""" initialize nbr module """

# NOTE: make sure there is only 1 .nbr-file in the active domain!!
# if no or multiple .nbr-files are present, CFPy throws an error

nbr = cfpy.nbr()

# read model layer/node and conduit layer/node elevations

# bot_elev has shape (n_layers, n_rows, n_cols)
# cond_elev has shape (n_conduit_layers, n_rows, n_cols)
# function looks for an .nbr-file, if none or more than one file is found, an error is raised
# make sure to only have one .nbr-file in the working directory
bot_elev, cond_elev = nbr.nbr_read()

# create nbr-data
nbr_data = nbr.nbr(bot_elev, cond_elev)
```

```python
# nbr-data contains:
#     0: node numbers
#     1: plane numbers
#     2: node locations
#     3: conduit locations
#     4: node neighbors
#     5: tube numbers
#     6: tube pairs
#     7: tube neighbors
```

```python
# nbr data: locations in grid are indexed differently:
#     MODFLOW: (LAY, ROW, COL)
#     CFPy: (COL, ROW, LAY)
```

## Plot Network

```python
network = cfpy.plots.Network(elevs=[bot_elev, cond_elev], nbr_data=nbr_data)
plot = network.plot_network(text_shift=0.1, dpi=70, rot_x=15, rot_z=110, plot_nums=True, kind="rectangular")
```

### CFPy parametrization of conduits

```python
# make pipe parametrization lists

# get pipe numbers
p_num = nbr_data[5]

# parametrization of all k pipes
# assuming same characteristics for all pipes
# lists for all conduit parametrization can be accessed individually to make differently parametrized pipes
# e.g., p_diameter[13] = 0.45

# diameter
p_diameter = np.ones(len(p_num)) * 1.0
p_diameter = p_diameter.tolist()

# tortuosity
p_tortuosity = np.ones(len(p_num)) * 1.0
p_tortuosity = p_tortuosity.tolist()

# roughness height
p_rheight = np.ones(len(p_num)) * 0.05
p_rheight = p_rheight.tolist()

# lower critical reynolds number
p_lcritrey = np.ones(len(p_num)) * 500
p_lcritrey = p_lcritrey.tolist()

# higher critical reynolds number
p_hcritrey = np.ones(len(p_num)) * 5000
p_hcritrey = p_hcritrey.tolist()

# summarize all data in a 2D-array
pipe_data = [p_num, p_diameter, p_tortuosity, p_rheight, p_lcritrey, p_hcritrey]
```

```python
# make lists for node head boundary conditions and tube-matrix exchange parameters

# get node numbers
n_num = nbr_data[0]

# parametrization for all n nodes
# if head > 0: fixed head
# if head == -1: calculated head
n_head = np.ones((len(n_num))) * -1
n_head = n_head.tolist()

# set individual node boundary conditions here
# e.g., for n_head[-1] the last node has a constant head given by the variable node_head_bc
n_head[20] = node_head_bc

# conduit wall conductivity in [m/d]
# identical for all conduits
# create list containing individual values to parametrize conduits individually
kex = 1.

# change units
if time_unit == 3:
    k_exchange = np.ones(len(n_num)) * kex / 24
    k_exchange = k_exchange.tolist()
if time_unit == 4:
    k_exchange = np.ones(len(n_num)) * kex
    k_exchange = k_exchange.tolist()

# summarize parametrization in lists
node_data = [n_num, n_head]
kex_data = [n_num, k_exchange]

# parametrization of conduit associated storage (CADS) in [m]
# again assuming identical values for all conduits
cads_data = np.ones((len(n_num))) * 0.05
cads_data = cads_data.tolist()
```

### CFPy - CFP

```python
""" initialize CFP module """

# if CADS should not be used or standard CFP (CFPv1) is used set cads = None below!

cfp = cfpy.cfp(mode = 1, # mode 1 == conduits, 2 == preferential flow layer (not supported), 3 == both (not supported)
               nnodes = len(n_num), 
               npipes = len(p_num), 
               nlay = n_lays, 
               nbr_data = nbr_data, 
               geoheight = cond_elev, 
               sa_exchange = 1, 
               epsilon = cfptol,
               niter = 1000, 
               relax = cfprelax, 
               p_nr = 0, 
               cond_data = pipe_data, 
               n_head = node_data, 
               k_exchange = kex_data, 
               ncl = 0, 
               cl = 0, 
               ltemp = 10,
               condl_data = 0,
               cads = cads_data)

# create list of strings
cfp = cfp.cfp()
```

### CFPy - COC

```python
""" initialize COC module """

coc = cfpy.coc(nnodes = len(nbr_data[0]),
               node_numbers = nbr_data[0],
               n_nts = 1,
               npipes = len(nbr_data[5]), 
               pipe_numbers = nbr_data[5],
               t_nts = 1)

# create list of strings
coc = coc.coc()
```

### CFPy - CRCH

```python
""" initialize CRCH module """

# p_crch is list of length (n_nodes) specifying the direct recharge fraction for each node
# initially set p_crch to 0 for all nodes (no direct recharge)
p_crch = np.zeros((len(n_num))).tolist()

# input direct recharge fraction (0 <= p_crch <= 1) for individual nodes
p_crch[0] = 0.1
p_crch[10] = 0.1

crch = cfpy.crch(iflag_crch = 1, 
                 nper = n_pers, 
                 node_numbers = nbr_data[0], 
                 p_crch = p_crch)

# create list of strings
crch = crch.crch()
```

# MODFLOW Input Files (FloPy)
### DIS

```python
""" DIS (Discretization): Creates model grid """

# Model grid:
# Note: MODFLOW defaults to units of m/day
# overall dimensions of domain (in x- and y-direction)
Lx = n_cols*delr
Ly = n_rows*delc
mtop = np.array(bot_elev[0][:][:])
mbot = np.array(bot_elev[1:][:][:])

# Create dis object:
dis = flopy.modflow.ModflowDis(mf, n_lays, n_rows, n_cols, n_pers, delr, delc, top=mtop, botm=mbot,
                               perlen=perlen, nstp=n_stps, steady=steady, itmuni=time_unit, lenuni=2)

# itmuni: 4 - days, 3 - hours 
```

### BAS

```python
""" BAS6 (Basic): Assigns constant head boundaries """

# if ibound < 0: constant head, if = 0: inactive, if > 0: active.
# create arrays to indicate active cells (ibound) and starting heads (H_init)

# integer array of dim (z,y,x), makes all cells active 
ibound = np.ones((n_lays, n_rows, n_cols), dtype=np.int32)

# replace leftmost column with -1 to indicate constant head 
ibound[:,:,0] = -1

# creates linearly spaced head values to use along rows
H_rows = np.linspace(H_left, H_right, n_cols)
# initialize starting head array
H_init = np.ones((n_lays, n_rows, n_cols), dtype=np.float32)
# assign starting head values to all cells
H_init[:,:,:] = H_rows

bas = flopy.modflow.ModflowBas(mf, ibound=ibound, strt=H_init)
```

### LPF

```python
""" LPF (Layer Property Flow): Assigns horizontal & vertical flow parameters between cells """
# Only one of either the BCF, LPF, or HUF2 packages should be used to control flow between cells
# Can also in this package set confined/unconfined (defaults to confined), cell-by-cell Ks as an array, 
# storage, specific yield, etc.
# defaults to confined

#unit number to save outputs to
ipakcb = 50
#value assigned to dry cells
hdry = 999.
#0=confined, >0 = unconfined/convertible
laytyp = 1
#method of calculating interblock tranmissivity, 0=harmonic mean (default)
layavg = 0
#horizontal anisotropy (ratio of Ky (along cols) to Kx (along rows)
chani = 1
#indicates vka = Kz
layvka = 0
#wetting, 0=inactive, 1=active
laywet = 0

lpf = flopy.modflow.ModflowLpf(mf, laytyp=laytyp, chani=chani, layvka=layvka, laywet=laywet, 
            ipakcb=ipakcb, hdry=hdry, hk=hk, hani=chani, vka=vka, ss=ss, sy=sy, wetdry=laywet) 
```

### OC

```python
""" OC (Output Control): Decides what outputs to save """

# This rarely needs to be changed

# dictionary for stress period data, where key (0,0) is a tuple with
# the zero-based period and step for each print/save, e.g.:
# spd = {(0,0):['print head', 'save head', 'save budget'],
#       (1,0):['print head', 'save head', 'save budget'],
#       (2,0):['print head', 'save head', 'save budget'],
#       (3,0):['print head', 'save head', 'save budget'],
#       (4,0):['print head', 'save head', 'save budget']}

# zero based list of stress period numbers
per_range = range(n_pers)
# empty list for storing tuples describing (stress per. number, time step) for printing / saving sim. results
store_per_tuple = []
# create tuples (always printing on first time step in stress period)
for index, (per, ts) in enumerate(zip(per_range, np.ones_like(per_range) * (n_stps - np.ones_like(n_stps)))):
    store_per_tuple.append(tuple((per, ts)))

# create dictionary
per_dict = {}
for per_tuple in store_per_tuple:
    per_dict[per_tuple] = ['print head', 'save head', 'save budget']

# create output control file using dictionary
oc = flopy.modflow.ModflowOc(mf, stress_period_data=per_dict)
```

### PCG

```python
""" PCG (Preconditioned Conjugate Gradient) """
# This rarely needs to be changed

#max number of outer iterations
mxiter = 500
#number of inner iterations
iter1 = 100
#flag for matrix conditioning method, 1=modified incomplete cholesky
npcond = 1
#flag for active cell surrounded by dry cells. 0=converts to dry, not0=converts only if hcof=0
ihcofadd = 9999
#head change criterion for convergence
hclose = mftol
#residual criterion for convergence
rclose = mftol
#relaxation parameter with npcond=1
relax = mfrelax
#estimate of upper bound
nbpol = 2
#printout interval
iprpcg = 5
#print flag, 5=?
mutpcg = 0
#damping factor
damppcg = 1.

pcg = flopy.modflow.ModflowPcg(mf, mxiter=mxiter, iter1=iter1, npcond=1, hclose=hclose, rclose=rclose, 
                               relax=relax, nbpol=nbpol, iprpcg=iprpcg, mutpcg=mutpcg, damp=damppcg,
                               ihcofadd=ihcofadd)
```

### RCH

```python
""" RCH: Recharge """
# recharge flux distributed over the top of the model - MODFLOW multiplies the fluxes by the horizontal area of the cells 
# to which they are applied to calculate the volumetric flux rates

# generate recharge data
np.random.seed(2)
rch_ts = np.random.gamma(0.05, 0.002, (n_pers))
plt.plot(rch_ts)

# flag, 1=rech only applied to top layer
nrchop = 1
# if non-zero, cell budget data will be saved
ipakcb = 50
# layer to which recharge is applied to (only used if nrchop=2)
irch = 1

# dictionary of recharge fluxes for each stress period
rech = {}
    
for num, rech_in in enumerate(rch_ts):
    
    # make array (n_rows x n_cols) to store cell-by-cell recharge information
    # apply recharge to all cells then apply direct recharge in relevant cells
    rech_data = np.ones((n_rows, n_cols)) * rech_in
    
    # direct recharge cells have to be given explicitly
    # such as
    # rech_data[ROW, COL] = some_direct_recharge
    
    rech[num] = rech_data

rch = flopy.modflow.mfrch.ModflowRch(mf, nrchop=nrchop, ipakcb=ipakcb, rech=rech, irch=irch)
```
